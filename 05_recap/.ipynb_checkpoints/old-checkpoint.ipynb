# 1 - Put it all together: SIMD + OpenMP + MPI

Computer memory is organized in a hierarchy of decreasing price and speed and increasing capacity. Caches are small and fastand RAM is big and slow. 

So chaches need to map memory several memory locations (asa juggler does with balls)

Memory allocation in the heap is managed by functions such:
- C: `malloc`, `calloc`, `free`.  
- C++: new, delete.  
- Fortran: ALLOCATE.  

If available, a pointer to memory reserved is delivered by the OS.

Programs often re-use nearby memory locations:
- Temporal locality: probably we will need again soon data from a location
- Spatial locality: nearby addresses likely to be needed soon

Once we have set up a decent piece of code we can ask the copiler to help a little bit:

## Vectorization

- Perform one operation on multiple elements of a vector  
- Chunk-wise processing instead of element wise  
- Can lower computational cost  
- A form of **Single Instruction Multiple Data** (SIMD) parallelism

basically, in addition to adding cores, I also add more FPU units per core

```python
Image("pictures/inst_set.png")
```

Vectorizationis handled by *compiler* switches or (from version 4.0 onwards) by *OpenMP pragmas* i. e. by a compiler directive (not a part of the language used) that instructs the compiler how to genrate the code that follows:

    #pragma omp simd <something>
    !$OMP simd <something>

```python
%%writefile Makefile

CC=gcc
CFLAGS= -std=c99 -fPIC -pedantic -Wall -fopenmp
OPTFLAGS=-O3 -ffast-math -march=native -mavx2 -fopenmp-simd -DOPT -fopt-info-vec-optimized
LDFLAGS=-lm -fopenmp
SHELL   = /bin/sh
OBJECTS  = mandelbrot.o
TARGET   = mandelbrot.exe

all: $(TARGET) 

opt: CFLAGS += $(OPTFLAGS) 
opt: CXXFLAGS += $(OPTFLAGS)
opt: $(TARGET)

$(TARGET): $(OBJECTS)
	$(CC) -o $(TARGET) $(OBJECTS) $(LDFLAGS)
	
%.o : %.c
	$(CC) -c $(CFLAGS) $(CPPFLAGS) $< -o $@

.PHONY: clean	
clean:
	rm -f $(TARGET) $(OBJECTS)
```

```python
%%writefile serial_0.c
// g mancini apr 18
#include <stdio.h>
#include <stdlib.h>
#include <math.h>
#include <string.h>
#include <omp.h>

    int mandel(double x0, double y0, double tr, int maxiter)
    {
        double re  = 0.0;
        double im  = 0.0;
        double re2 = 0.0;
        double im2 = 0.0;

        for (int k = 1; k < maxiter && (re2 + im2 < tr); k++) 
        {
            im = 2. * re * im + y0;
            re = re2 - im2 + x0;
            re2 = re * re;
            im2 = im * im;
        }    
        return k;
    }

void write_output(char* filename, int* image, int width, int height, int maxiter)
{
    
    // open output PPM file, https://en.wikipedia.org/wiki/Netpbm_format#PPM_example
    unsigned char colour[3];
    FILE * fp = fopen(filename,"w");
    fprintf(fp,"P3\n%d %d %d\n",width, height, 255);    

    for (int i=0; i< height; i++)
        for (int j=0; j < width; j++)
        {           
            if(image[i + j*height] >= maxiter)
            {
                colour[0] = 0;
                colour[1] = 0;
                colour[2] = 0;
            }
            else
            {
                colour[0] = 255;
                colour[1] = 255;
                colour[2] = 255;
            }
            fprintf(fp,"%d %d %d ",colour[0],colour[1],colour[2]);
        }

    fclose(fp);     
}
    
int main(int argc, char *argv[])
{

    const int maxiter   = 1000;
    double horiz  = 3.0;
    const int width     = 2048;
    const int height    = 2048;
    const double xmin   = -2.;
    const double xmax   =  1.;
    const double ymin   = -1.2;
    const double ymax   =  1.2;
    char* filename = "out.ppm";
    
    horiz = horiz*horiz;
    double x0, y0, xres, yres;
    xres = (xmax - xmin)/((double) width);
    yres = (ymax - ymin)/((double) height);
    
    int *image = malloc(sizeof(int*) * height * width); 
    
    double t0 = omp_get_wtime();
    y0 = ymin;
    for (int i=0; i< height; i++)
    {
        x0 = xmin;
        for (int j=0; j < width; j++)
        {
            image[i + j*height] = mandel(x0,y0,horiz,maxiter);
            x0 = x0 + xres;
        }   
        y0 = y0 + yres;
    }// close on i    
    printf("Spent %f seconds\n",omp_get_wtime()-t0);
    
    write_output( filename, image, width, height, maxiter);
    free(image);
}
```

## The zeroth law of optimization

### "Premature optmization is the root of all evil" - Donald Knuth

### The wisdom of Bjarne Stostrup

I get maybe two dozen requests for help with some sort of programming or design problem every day. Most have more sense than to send me hundreds of lines of code. If they do, I ask them to find the smallest example that exhibits the problem and send me that. Mostly, they then find the error themselves. *Finding the smallest program that demonstrates the error* is a powerful debugging tool.
