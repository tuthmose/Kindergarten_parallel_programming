Let's write a simple C code to estimate $\pi$ and implement it in **C** and **Fortran**. We estimate $\pi$ as:
$$ \int_0^1 \frac {1} {1+x^2} = arctg(1) - arctg(0) = \pi/4 $$

```python
%%writefile simple_pi.c
#include <stdlib.h>
#include <stdio.h>
#include <math.h>

int main (int argc, char **argv) {
        

    int num_steps = argv[1] != NULL ? atoi(argv[1]) : 1E8;

    double x, PI, sum=.0;

    for (int i = 0; i < num_steps; i++){
        x = (i+0.5)/((double) num_steps);
        sum = sum + 4.0/(1.0+x*x);
    }
    PI = sum / ((double) num_steps);

    printf("PI: %f\n",PI);
    return 0;
}
```

### Memory Usage

Remember that computer memory is organized in a hierarchy of decreasing price and speed and increasing capacity (source: Wikipedia)

```python
Image("bulldozer.png")
```

1. *Heap*:
    - Large allocations.
    - Needs management.

2. *Stack*:
    - First in, Last out (“FILO”) type.
    - Short term memory

Memory allocation in the heap is managed by functions such:
- C: `malloc`, `calloc`, `free` or `_mm_malloc` (or masked by some higher level interface).  
- C++: new, delete.  
- Fortran: ALLOCATE.  

If available, a pointer to memory reserved is delivered by the OS.

Programs often re-use nearby memory locations:
- Temporal locality: probably we will need again soon data from a location
- Spatial locality: nearby addresses likely to be needed soon
    
That's why we have caches.

Cache divided into cache lines: when memory is read into cache, a whole cache line is always read at the same time.  
Good if we have data locality: nearby memory accesses will be fast
Typical cache line size: 64-128 bytes

How caches work:

**Direct-mapped cache**:
- Hash function maps to a single place for every unique address.  

**Set associative cache**:  
- Cache space is divided into n sets – Address is distributed modulo n – k-way set associative (k = 2...24).  

**Fully associative cache**:  
- Maps to any slot.  
    
In practice, k-way set associative commonly used, e.g. L1d cache commonly 2-way set associative.

*Cache hit* 
*Cache miss*

Reasons for a miss:
- Compulsory: On the first access to a block.
- Capacity: When blocks are being discarded from cache because cache cannot contain all blocks.
- Conflict: In set associative or direct mapped strategies, conflict misses occur when several blocks are mapped to the same set or block frame

### Knowing your compiler

You have been told that GCC have several optimization levels (`-On`):

- -O0 : Reduce compilation time and make debugging produce the expected results. This is the default.
- -O or -O1 : Optimize. The compiler tries to reduce code size and execution time, without performing any optimizations that take a great deal of compilation time.
- -O2 Performs nearly all supported optimizations that do not involve a space-speed tradeoff. This is somewhat the *optmistic default* and it's used in many instances (Makefile, automated building processess). 
- -Os : optimise for size

You can inspect what `-O1` should really means:

```python
gcc -O1 -Q --help=optimizers | grep enabled | wc -l
```

```python
gcc -O1 -Q --help=optimizers | grep enabled | grep inline
```

Why so many options? Because optimization is **very complex**: for each high level code snippet there **many** possible low level implementations. Developers and testers need a very fine grained control on what the compiler tries to do.

Two *very simple* source code level optimizations that a compiler will attempt are **common subexpression optimization** and **function inlining**.

**CSE** is about re-using computed values, especially for costly functions, by creating temporary variables:
    
    a = cos(v)*(1+sin(u) + cos(w)*(1-sin(u))
    
evaluates `sin(u)` twice. However:

    double temp = sin(u)
    a = cos(v)*(1+temp)+cos(w)*(1-temp)
    
only does it once. The tradeoff between creating a new variable and evaluating a trigonometric functions is usually favorable.

**inlining** tries to reduce function call overhead for very small functions, such as:

    int double_it(int x)
    {
        return x+x;
    }
    for(int i=0; i<1E6; i++)
        a += double_it(i*3);
        
here, the cost of calling from memory the function `double_it` exceeds that of the computation. Again, the compiler will make use of temporary variables:

    for(int i=0; i<1E6; i++)
    {
        int temp = i*6
        a += temp
    }

as you can see from above `-O1` involves function inlining under different conditions

Another option activated by `-O1` is `-fmove-loop-invariants` which will perform something like that (from Wikipedia):

    for (int i = 0; i < n; i++) {
        x = y + z;
        a[i] = 6 * i + x * x;
    }

    x = y + z;
    t1 = x * x;
    for (int i = 0; i < n; i++) 
        a[i] = 6 * i + t1;

While CSE reduces both the size and runtime of the code, inlining can achieve better performance at the cost of increased executable size. This is a *space time tradeoff*.  
Another common source code optimization that a compiler will attempt is **loop unrolling**:

    for(int i=0; i<10; i++)
        a[i] = i+1;
        
the loop contains very few iterations and there is a overhead for managing the loop. The compiler then does:

    a[0] = 1;
    a[1] = 2;
    ...
    a[9] = 10;
    
even if for a human this would less readable. 

Use the `const` keyword to tell the compiler that the a value is read only.  
Allows compiler to optimize more because the compiler knows more.  
Using `const` also helps avoid programming errors, like modifying something by mistake.

Many source level optimizations, automatic or not, aims at:
- reduce the number of operations
- reduce their mathematical complexity
- increase the data locality of operations. This is perhaps the important thing

Automatic optimizations may be enhanced or inhibited by good coding even if nowadays we don't need to CSE by yourself

After source level optimizations, the compiler will deal the way to combine single instructions in different order (*scheduling*) or in parallel (*pipelining*).

Let's see what changes between `-O0` and `-O1`.

```python
%%writefile fibonacci.c
#include <stdio.h>
#include <stdlib.h>

void fib(unsigned *seq, unsigned position)
{
    seq[position] = seq[position-1]+seq[position-2];
    //return seq[position];
}

int main(int argc, char **argv)
{
    const unsigned L = argv[1] != NULL ? atoi(argv[1]) : 1E8; 
    unsigned  *F = calloc(L,sizeof(unsigned));
    
    F[0] = 0;
    F[1] = 1;
    
    if(L>1)
        for (unsigned i = 2; i <= L; i++)
        {
            fib(F,i);
        }

  return F[L];
}
```

What if you want to get an estimate on how much time is spent in some part of the code?  
Quick and dirty: `omp_get_wtime()`

```python
%%writefile fibonacci.c
#include <stdio.h>
#include <stdlib.h>
#include <omp.h>

void fib(unsigned *seq, unsigned position)
{
    seq[position] = seq[position-1]+seq[position-2];
    //return seq[position];
}

int main(int argc, char **argv)
{
    const unsigned L = argv[1] != NULL ? atoi(argv[1]) : 1E8; 
    unsigned  *F = calloc(L,sizeof(unsigned));
    
    F[0] = 0;
    F[1] = 1;
    
    if(L>1)
    {
        double t0 = omp_get_wtime();
        for (unsigned i = 2; i <= L; i++)
        {
            fib(F,i);
        }
        printf("spent in loop %f\n",omp_get_wtime()-t0);        
    }
  return F[L];
}
```

### Exercises

Consider the following:
    
$$ A^{nxn} = A^{nxn} + B^{nxn}*C^{nxn}$$

a naive C implementation may look like that:

```python
%%writefile matmul_dot.c
void matmul(int **a, int **b, int **c, int dim)
{
    for (int i = 0;i < dim; i++)
        for (int j = 0; j < dim; j++)                
            for (int k = 0;k < dim; k++)                 
                a[i][j] += b[i][k] * c[k][j];  
}
```

```python
%%writefile matmul_test.c
#include "matmul_test.h"

int main(int argc, char **argv)
{
    int n = argv[1] != NULL ? atoi(argv[1]) : 1000; 
    int high = argv[1] != NULL ? atoi(argv[2]) : 100; 

    srand(time(0));
    //in practice, do not forget to check for NULL
    int **A = malloc(n * sizeof(int*));
    for(int i = 0; i < n; i++)
        A[i] = malloc(n * sizeof(int));
    
    int **B = malloc(n * sizeof(int*));
    for(int i = 0; i < n; i++)
        B[i] = malloc(n * sizeof(int));
    
    int **C = malloc(n * sizeof(int*));
    for(int i = 0; i < n; i++)
        C[i] = malloc(n * sizeof(int));
    
    matset(A,n,high);
    matset(B,n,high);
    matset(C,n,high);
    if(n == 3)
    {
        matprint(A,n);
        matprint(B,n);
        matprint(C,n);
    }

    matmul(A,B,C,n);
    
    if(n == 3)
        matprint(A,n);
    return 0;
}
```

```python
%%writefile matset.c
#include "matmul_test.h"

void matset(int **a, int dim, int high)
{
   
    for (int i = 0; i < dim; i++)
        for(int j = 0; j < dim; j++)
                a[i][j] = rand() % high+1;
}
```

```python
%%writefile matprint.c
#include "matmul_test.h"

void matprint(int **a, int dim)
{
    for(int i=0; i < dim; i++)
            printf("%d %d %d\n",a[i][0],a[i][1],a[i][2]);
}
```

#### part 1

```python
%%writefile matmul_test.c
#include "matmul_test.h"
#include <omp.h>

int main(int argc, char **argv)
{
    int n = argv[1] != NULL ? atoi(argv[1]) : 1000; 
    int high = argv[1] != NULL ? atoi(argv[2]) : 100; 

    srand(time(0));
    //in practice, do not forget to check for NULL
    
    double t0 = omp_get_wtime();
    int **A = malloc(n * sizeof(int*));
    for(int i = 0; i < n; i++)
        A[i] = malloc(n * sizeof(int));
    
    int **B = malloc(n * sizeof(int*));
    for(int i = 0; i < n; i++)
        B[i] = malloc(n * sizeof(int));
    
    int **C = malloc(n * sizeof(int*));
    for(int i = 0; i < n; i++)
        C[i] = malloc(n * sizeof(int));

    double t1 = omp_get_wtime();
    printf("spent in allocation %f\n",t1-t0);   
    
    matset(A,n,high);
    matset(B,n,high);
    matset(C,n,high);
    double t2 = omp_get_wtime();
    printf("spent in set %f\n",t2-t1);
    
    if(n <= 3)
    {
        matprint(A,n);
        matprint(B,n);
        matprint(C,n);
    }

    matmul(A,B,C,n);
    double t3 = omp_get_wtime();
    printf("spent in multiplication %f\n",t3-t2);    
    
    if(n <= 3)
        matprint(A,n);
    return 0;
}
```

#### part 4

We spent a lot of time in class 1 speaking about caches and memory bottlenecks.  
In the way `matmul` is written, each element of A is the dot product of a row of B and column of C.  
If we swap the loops on `j` and `k` then, for each row of A (loaded in cache), we use one element in B and the same row in C:

```python
%%writefile matmul_blas.c
void matmul(int **a, int **b, int **c, int dim)
{
    for (int i = 0;i < dim; i++)
        for (int k = 0;k < dim; k++)        
            for (int j = 0; j < dim; j++)                               
                a[i][j] += b[i][k] * c[k][j];  
} 
```

### (GNU) Makefiles

Optimization, warnings, path and other options make very difficult to compile large projects from the command line. That's why there a lot of tools of automated/assisted build.

E. g. And IDE like Qtcreator creates template and project (*.pro) files

some other tools for automatic creation of customized enviroment are:
- CMake
- Scons
- GNU Autotools

and many others

The simplest (but not most efficent) tool is the humble `Makefile` containing shell commands to run the compilation and linking processes.  
Make keeps track of the last time files (normally object files) were updated and only updates those files which are required (ones containing changes) to keep the sourcefile up-to-date. 

The makefile contains a list of **rules**. These tell the system what commands you want to be executed.  
The rules, which must begin in column 1, are in two parts. The first line is called a **dependency line** and the subsequent line(s) are called actions or commands. The action line(s) must be indented with a tab.  
   
    RULE:	DEPENDENCY LINE
    [tab]ACTION LINE(S)

The dependency line is made of two parts. The first part (before the colon) are target files and the second part (after the colon) are called source files. It is called a dependency line because the first part depends on the second part. Multiple target files must be separated by a space. Multiple source files must also be separated by a space.

    DEPENDENCY LINE: TARGET FILES:	SOURCE FILES

## What is vectorization?

- Perform one operation on multiple elements of a vector  
- Chunk-wise processing instead of element wise  
- Can lower computational cost  
- A form of **Single Instruction Multiple Data** (SIMD) parallelism

Think of vectorization in terms of loop unrolling:

    for(int i = 0;i < max; i++)
        A[i] = B[i] + C[i]*D[i]

### Finer control: pragmas

Limits on autovectorization:
    
- Data dependencies
- Function calls in loop block
- Complex control flow / conditional branches
- Loop not “countable”

OpenMP 4.0 offers support for a standard implementation of vector instructions using [pragmas](#https://en.wikipedia.org/wiki/Directive_(programming) which are managed by the compiler. As an example, [GCC](#https://gcc.gnu.org/onlinedocs/gcc/Pragmas.html) supports many pragmas for different purposes.
Some set of pragmas (such as those defined in the OpenMP standard) are supported by many compilers (see [here](#https://gcc.gnu.org/onlinedocs/libgomp/)).

In particular, it is possible to define pragmas to vectorize a specific loop construct, ignoring others:
    
    \#pragma omp simd 
     for-loop
     
    !$omp simd 
    do-loop
    !$omp end simd

    void matmul(int *a, int *b, int *c,int n) {
        for (int i=0; i<n; i++)
            for (int k=0; k<n; k++)
                #pragma omp simd 
                    for (int k=0; k<n; k++)
                    a[i][j] += b[i][k] * c[k][j];

SIMD pragmas are activated at `-O2` by specifying `-fopenmp-simd`:

```python
%%writefile loop_simd.c
void sum(int *a, int *b, int *c, int size)
{   
    #pragma omp simd 
    for(int i=1; i < size; i++)
        a[i] += b[i-1];    

    for(int i=1; i < size; i++)
        a[i] += c[i];       
}
```

```python
gcc -Wall -std=c99 -O2 -fopt-info-vec-optimized -march=native -mssse3 -fopenmp-simd -c -o loop_simd.o loop_simd.c
```

Additional settings in a pragma directive can be set with *clauses*:

- `safelen (length)`
    - Maximum number of iterations that can run concurrently without breaking a dependence

- `simdlen (length)`
    - Specify preferred length of SIMD registers used

- `linear (list[:linear-step])`
    - The variable’s value is in relationship with the iteration number: `x i = x orig + i * linear-step`
    
- `ivdep`
    - No vector dependence. Alternative to `__restrict`

#### Data sharing clauses

A particularly important category of clauses are those that perform *data sharing* between *vector lanes*.  
Consider the following snippet

    #pragma omp simd
    for(int i=0; i<n; i++)
        sum += a[i];
        
can we vectorize it? how?

one way would be to create copies of the`sum` variable and store partial sums (*promoting* the scalar to vector):

    sum[2] = {0.;0.};
    #pragma omp simd
    for(int i=0; i<n; i=i+2)
        sum[0] += a[i];
        sum[1] += a[i+1];
        
and then collect the results (*reducing* again the vector to a scalar):

    sum_final = sum[0] + sum[1]

This is such a common need that the *reduction* is supported (for simple operators) by the standard:
    
    #pragma omp simd reduction(+:sum)
    for(int i=0; i<n; i++)
        sum += a[i];

#### SIMD Function Vectorization

define, in one library, use somewhere esle

    #pragma omp declare simd
    float distsq(float x, float y) 
    {
        return (x - y) * (x - y);
    }

    void example() 
    {
        #pragma omp parallel for simd
        for (i=0; i<N; i++) 
            d[i] = min(distsq(a[i], b[i]), c[i]);

    }

Even for quite complex functions:

    #pragma omp declare simd
    float MyErfElemental(const float inx)
    {
        const float x = fabsf(inx); // Absolute value (in each vector lane)
        const float p = 0.3275911f; // Constant parameter across vector lanes
        const float t = 1.0f/(1.0f+p*x); // Expression in each vector lanes
        const float l2e = 1.442695040f; // log2f(expf(1.0f))
        const float e = exp2f(-x*x*l2e); // Transcendental in each vector lane
        float res = -1.453152027f + 1.061405429f*t; // Computing a polynomial
        res = 1.421413741f + t*res; // in each vector lane
        res =-0.284496736f + t*res;
        res = 0.254829592f + t*res;
        res *= e;
        res = 1.0f - t*res; // Analytic approximation in each vector lane
        return copysignf(res, inx); // Copy sign in each vector lane
    }

#### Vector dependence

Sometimes instructions in loops create data dependencies that limit vectorization.  
Look at the following loop:

    #pragma omp simd safelen(2)
    for( i=1; i<N; i++)
         a[i] = a[i-1] + b[i];

given `a={0,1,2,3,4,5}`, `b{1,2,3,4,5,6}` at the end `a[5] = 31`.  
This is a *write after read* situation. What would be the result with a vectorization lenght of `2`?

values in block are loaded in memory:
    
load:    a[1-1,2-1] = {0,1}, b[1,2] = {1,2}  
operate: {0,1} + {2,3} = {2,4}   
store:   a[1,2] = {2,4}  

load:    a[3-1,4-1] = {4,3}, b[3,4] = {3,4}   
operate: {4,3} + {3,4} = {7,7}   
store:   a[3,4] = {7,7}  

load:    a[4-1,5-1] = {7,5}, b[4,5] = {5,6}  
operate: {7,5} + {5,6} = {16,11}   
store:   a[4,5] = {16,11}  

What about a `read after write`?

    #pragma omp simd safelen(4)
    for( i=0; i<N-1; i++)
     a[i] = a[i+1] + b[i];

load:    a[i+1] = {1,2,3,4}, b[i] = {1,2,3,4}   
operate: {1,2,3,4} + {1,2,3,4} = {2,4,6,8}   
store:   a[i] = {2,4,6,8}  

load:    a[i+1] = {4,5}, b[i] = {5,6}  
operate: {4,5} + {5,6} = {9,11}   
store:   a[i] = {9,11}  

What about:
    
    for (int i = m; i < n; i++)
        a[i] += a[i-m]; 

Vector dependence can be caused by aliasing. To let the compiler know that there's no aliasing:

```python
%%writefile loop_simd.c
void sum(int *a, int *b, int *c, int size)
{   
    #pragma omp simd 
    for(int i=1; i < size; i++)
        a[i] += b[i-1];    

    for(int i=1; i < size; i++)
        a[i] += c[i];       
}
```

```python
%%writefile integral.c
#include "integral.h"

int main(int argc, char **argv)
{
    double a  = atof(argv[1]);
    double b  = atof(argv[2]);
    int n = atoi(argv[3]);
    if(a >= b)
    {
        printf("error: b must be > a\n");
        return -1;
    }
    
    const double dx = (b - a)/n;
    double I = 0.;
    double t0 = omp_get_wtime();    

    for (int i = 0; i < n; i++)
    {
        const double xip12 = a + dx * ((double ) i + 0.5);
        const double yip12 = myfunc(xip12);
        const double dI = yip12 * dx;
        I += dI;
    }
    printf("Loop completed in %g s.\n Integral value=%f\n",omp_get_wtime()-t0,I);
    
}
```

Let's try to create Python versions of the $\pi$ program and compare results:

1) A *pure Python* version:

```python
def simple_pi(num_iter):
    mysum  = .0
    step = 1./num_iter
    
    for i in range(num_iter):
        x = (i+0.5)*step
        mysum += 4./(x*x + 1)
        
    #print(mysum*step)
    return mysum*step
```

2) A *numpy* version:

```python
def numpy_pi(num_iter):
    grid = np.linspace(0,1,num=num_iter,endpoint=False)
    x    = 4./(1. + grid**2)
    pi   =  (1./num_iter)*np.sum(x)
    return pi
```

## Why Python is slow?

If you like to be pendantic, we should remember that Python is a *standard* with *different implementations* as the  [Python web site](#https://wiki.python.org/moin/Why%20is%20Python%20slower%20than%20the%20xxx%20language) remembers us. 

But then, *why running a for loop in CPython is slower than C?*

### Duck typing

Take the difference of C variable vs a Python one at runtime: 

- In C/C++, the values are stored directly in the memory block associated to a variable. Furthermore, since these languages are statically typed, each variable has a specific type which must be known at compile time. The compiler uses types information to allocate the proper amount of memory for each variable and to perform checks on the conversions,   operations etc. performed on variables.  
- In Python variables are typeless pointers referring an object. Type information are associated to the objects, not to variables. Checks on the conversions, applied operations etc. are performed at runtime by the interpreter.  

```python
Image(filename="box.png")
```

When the code executes, the Python interpreter decides what properties the variable may have depending *on context*. The following C code:

    int i,j,k;
    i = 2;
    j = 2;
    k = i*j;
    return k
    
means the CPU will use approximately seven cycles. 

The Python equivalent:

    i = 2
    j = 2
    k = i*j
    return k
    
involves the following operations on the **PyObject** (see below) C structures:

- creating the objects i, j and k
- assigning the appropriate attibutes
- selecting the method for the '\*' operator
- calculating the result and the appropriate type for k
  
and each of these operations means that the interpreter creates or uses something like that: 

    typedef struct _object {
    _PyObject_HEAD_EXTRA
    Py_ssize_t ob_refcnt;
    struct _typeobject *ob_type;
    } PyObject;

### Bytecode

From Wikipedia:

Bytecode, also known as p-code (portable code), is a form of instruction set designed for efficient execution by a software interpreter. Unlike human-readable source code, bytecodes are compact numeric codes, constants, and references (normally numeric addresses) which encode the result of parsing and semantic analysis of things like type, scope, and nesting depths of program objects. They therefore allow much better performance than direct interpretation of source code.

### Interpreted language

Again, from Wikipedia, an interpreter may be a program that either
1. executes the source code directly
2. translates source code into some efficient intermediate representation (code) and immediately executes this
3. explicitly executes stored precompiled code made by a compiler which is part of the interpreter system

Note that, in principle, almost any language can be *compiled* or *interpreted*, it depends on whether you have (or write) a compiler program or an interpreter one. What does the **python** (using CPython on Linux) command do when you execute `python somescript.py` ?

### Locality

One of the key features of Numpy is the locality of data that allows to access to rows and columns as C (or Fortran) plus the decoration that make possible all the fancy stuff. The values in the array are **contiguous** and have a common size.

```python
Image(filename="array.png")
```

On the opposite, a list object in CPython is represented by the following C structure. **ob_item** is an array of pointers to the list elements. **allocated** is the number of slots allocated in memory

    typedef struct {
	    PyObject_VAR_HEAD
	    PyObject **ob_item;
	    Py_ssize_t allocated;
	} PyListObject;

each element in the Python list is found by a pointer to a buffer of pointers, each of which points to a Python object which encapsulates its own data. Thus, operations like *append* or *pop* are cheap but running over all the elements in the list can be costly since it involves a great deal of referencing operations and (likely) a lot of copies to and from the memory.
